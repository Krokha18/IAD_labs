# Завантажимо зображення

```python
img = plt.imread("polar_light.jpg")
plt.figure(figsize=(9,12))
plt.imshow(img)
```

```python
def img_to_pixels():
    return img.reshape((img.shape[0]*img.shape[1], 3))

def pixels_to_img(pixels):
    return pixels.reshape( img.shape )
```

# Переведемо зображення у список пікселів

```python
pixels = img_to_pixels()
pixels
```

# Проведемо спрощення кольорової гами, на основі методу K-Means

```python
%%time
# 64
from sklearn.cluster import KMeans, MiniBatchKMeans
#kmeans = KMeans(n_clusters=64, n_init=3)
kmeans = MiniBatchKMeans(n_clusters=64, n_init=10, batch_size=1024*32)
kmeans.fit(pixels)
clusters = kmeans.labels_

medians = np.array([np.median(pixels[clusters==i], axis=0).astype('int') for i in sorted(np.unique(clusters)) ])
plt.figure(figsize=(9,12))
plt.imshow( pixels_to_img(medians[clusters]).astype('int8') )
```
